```python
class ReadData:
    def __init__(self, filename, hasHeader=False):
        self.file = open(filename, newline='')
        self.reader = csv.reader(self.file, delimiter=CSV_DELIMITER)

        if hasHeader:
            next(self.reader)

    def readData(self, data):
        try:
            row = next(self.reader)
            self.fillData(data, row)
            return True
        except Exception as e:
            self.file.close()
            return False

    def fillData(self, data, row):
        for i in range(JOINT_NUM):
            data[i] = [float(x) for x in row[i].lstrip().split(' ')]
```

```python
def extractFeatures(infilename):
    data = [[] for i in range(JOINT_NUM)]
    reader = ReadData(infilename, hasHeader=True)

#     print('Starting Extraction of {}...'.format(infilename))
    skeletonFeatures = Features()
    while reader.readData(data):
#         print(data)
        skeletonFeatures.extractFeatures(data)

#     print('Ending Extraction ...')
    skeletonFeatures.endExtraction()

    (angles, velocities, frames) = skeletonFeatures.getFeatures()

    print('Angles Shape: {}'.format(angles.shape))
    print('Velocities Shape: {}'.format(velocities.shape))

    return np.concatenate((angles, velocities))
```

```python
%%time

with open(rootFolder + featuresFile, 'w', newline='') as outfile:
    for (_, _, filenames) in os.walk(rootFolder + '/skeleton'):
        writer = csv.writer(outfile, delimiter=CSV_DELIMITER)
        printProgress(0, len(filenames), prefix = 'Progress:', suffix = '0/{0}'.format(len(filenames)), bar_length = 50)

        for (index, f) in enumerate(filenames):
            videoFeatures = extractFeatures(rootFolder + '/skeleton/' + f)
            classname = f.split('A')[0].split('0')[-1]
            writer.writerow([f.split('.')[0]] + videoFeatures.tolist() + [int(classname)])
            printProgress(index+1, len(filenames), prefix = 'Progress:', suffix = '{0}/{1}'.format(index+1, len(filenames)), bar_length = 50)
```
